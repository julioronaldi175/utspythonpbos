```python
celcius = int(input("Celcius : "))
reamur = celcius * 0.8
fahrenheit = (celcius * 1.8) + 32
kelvin = celcius + 273,15
    
print("\nCelcius    : ", celcius, "°C")
print("Reamur     : ", reamur, "°R")
print("Fahrenheit : ", fahrenheit, "°F")
print("Kelvin     : ", kelvin, "K")
```

```python
sec = int(input("Masukkan Jumlah Detik : "))
days = sec/86400
temp1 = sec%86400

hours = temp1/3600
temp2 = temp1%3600

minutes = temp2/60
sec = temp2%60

print ("Hari  : ", round(days))
print ("Jam   : ", round(hours))
print ("Menit : ", round(minutes))
print ("Detik : ", round(sec))
```
